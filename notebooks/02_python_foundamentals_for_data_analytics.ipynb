# Part 1: Basic Revenue Math & Booleans
This section handles the initial financial calculations using the provided formula:

$$Final\_Revenue = (Revenue \times (1 + Tax\_Rate)) \times (1 - Discount\_Rate)$$

**Tasks:**
1. Compute revenues including tax.
2. Compute final revenues after tax and discount.
3. Create Boolean lists for filtering.

```python
revenues = [120, 250, 80, 310, 95]
tax_rate = 0.18
discount_rate = 0.10

taxed_revenues = [r * (1 + tax_rate) for r in revenues]

final_revenues = [r * (1 + tax_rate) * (1 - discount_rate) for r in revenues]

is_over_100 = [r > 100 for r in revenues]

is_between = [100 <= r <= 300 for r in revenues]

print(taxed_revenues)
print(final_revenues)
print(is_over_100)
print(is_between)
```

### Why do parentheses matter in the formula?

In mathematics and programming, we follow the **Order of Operations** (PEMDAS). Parentheses ensure that specific operations happen first.

* **Without them:** Python might multiply the tax rate by the discount rate before applying it to the revenue.
* **With them:** We force the computer to calculate the "Taxed Revenue" as a single total first, and then apply the "Discount" to that specific total. 

$$Final\_Revenue = (Revenue \times (1 + Tax\_Rate)) \times (1 - Discount\_Rate)$$

# Part 2: Manual Data Processing
In this section, we calculate statistics manually using loops to understand the underlying logic of data aggregation.

```python
sales = [120, 250, 80, 310, 95]

total_revenue = 0
for s in sales:
    total_revenue += s

average_revenue = total_revenue / len(sales)

max_value = sales[0]
for s in sales:
    if s > max_value:
        max_value = s

count_over_150 = len([s for s in sales if s > 150])

commissioned_sales = [s * 1.05 for s in sales]

print(total_revenue)
print(average_revenue)
print(max_value)
print(count_over_150)
print(commissioned_sales)
```

### Understanding Manual Loops
While Python has built-in functions like `sum()` and `max()`, writing them manually using `for` loops helps us understand how the computer "thinks." A loop visits every single item in a list one-by-one, keeping a "running tally" or comparing the "current value" to the "best so far" to find the maximum.

# Part 3: Customer Segmentation
Categorizing customers based on revenue thresholds.

```python
revenues = [120, 250, 80, 310, 95]

segments = ["Premium" if r > 250 else "Standard" if r > 100 else "Low" for r in revenues]

counts = {
    "Premium": segments.count("Premium"),
    "Standard": segments.count("Standard"),
    "Low": segments.count("Low")
}

print(segments)
print(counts)
```

### Why does the order of `elif` statements matter?

Python evaluates conditions from **top to bottom**. Once it finds a condition that is `True`, it stops checking the rest. 

* If we checked `revenue > 100` first, a "Premium" value of $310 would trigger the "Standard" label immediately and exit the logic.
* By checking the highest value (`> 250`) first, we ensure "Premium" customers are identified correctly before the remaining data is checked against lower thresholds.

# Part 4: Data Cleaning
Removing duplicate customer IDs to ensure data integrity.

```python
customer_ids = [1, 2, 3, 3, 4, 5, 5, 6]

unique_ids_set = set(customer_ids)

unique_ids_list = list(unique_ids_set)

length_before = len(customer_ids)
length_after = len(unique_ids_list)

print(unique_ids_list)
print(f"Before: {length_before}, After: {length_after}")
```

### Why are sets unordered?

Sets are designed for **speed and uniqueness**. They use a technical process called **Hashing**. Instead of keeping a numbered list (0, 1, 2...), the computer places items in memory "buckets" based on their value. This makes it instant to check if an ID already exists, but it loses the original sequence of the data.

# Part 5: Dictionaries
Using key-value pairs to manage specific customer profiles.

```python
customer = {
    "name": "Anna",
    "revenue": 250,
    "city": "Yerevan"
}

customer["segment"] = "Premium" if customer["revenue"] > 250 else "Standard"

customer["revenue"] = 300

del customer["city"]

for key, value in customer.items():
    print(f"{key}: {value}")
```

### Why are dictionaries useful for structured data?

Dictionaries act like digital labels. Unlike a list, where you have to remember that "Position 0 is the Name," a dictionary allows you to use **Keys**. Calling `customer["name"]` is much more readable and prevents errors when the data structure grows or changes.

# Part 6: Advanced List Comprehensions
Performing complex list transformations in a single line of code.

```python
revenues = [120, 250, 80, 310, 95]

taxed_revenues = [r * 0.9 for r in revenues]

high_revenues = [r for r in revenues if r > 100]

labels = ["High" if r > 200 else "Normal" for r in revenues]

print(taxed_revenues)
print(high_revenues)
print(labels)
```

### List Comprehension vs. Standard Loops

* **List Comprehensions:** Are best for simple transformations (like adding tax to every number). They are faster for the computer to process and take up only one line of code.
* **Standard Loops:** Are better for complex tasks where you need to perform multiple steps or write detailed comments for other developers.

# Part 7: Pandas DataFrame Analysis
Using industry-standard tools to organize data into tables.

```python
import pandas as pd

revenues = [120, 250, 80, 310, 95]
data = {"revenue": revenues}
df = pd.DataFrame(data)

df["revenue_after_tax"] = df["revenue"] * 0.9

def get_segment(r):
    if r > 250:
        return "Premium"
    elif r > 100:
        return "Standard"
    else:
        return "Low"

df["segment"] = df["revenue"].apply(get_segment)

df = df[df["revenue"] > 100]

df = df.drop("revenue_after_tax", axis=1)

print(df)
print(df.shape)
```

### Why is Pandas easier than manual loops?

Working with a **Pandas DataFrame** is superior for data analysis because:
1. **Vectorization:** You can multiply an entire column by a number in one line without writing a `for` loop.
2. **Alignment:** It keeps all your data (Revenue, ID, Segment) perfectly aligned in a table format.
3. **Efficiency:** Operations on DataFrames are written in C-code under the hood, making them much faster for large datasets.

# Bonus Reflection
1. Mutable vs. Immutable Objects
Mutable: Objects that can be changed after creation without changing their identity (e.g., Lists, Dictionaries, Sets). You can add or remove items in place.

Immutable: Objects that cannot be changed once created (e.g., Strings, Tuples, Integers). Any "change" actually creates a brand-new object in memory.

2. Loops vs. Pandas Vectorization
Loops are iterative; Python processes one row, moves to the next, and repeats. This involves high overhead for each step. Vectorization allows Pandas to perform operations on entire arrays at once by offloading the work to highly optimized C and Fortran code.

3. Boolean Logic and SQL WHERE
Boolean logic in Python (df[df["revenue"] > 100]) functions exactly like a SQL WHERE clause.

Python: df[df["price"] > 50]

SQL: SELECT * FROM table WHERE price > 50;
Both act as a "mask" or filter that only allows rows evaluating to True to pass through.

4. Lists vs. DataFrames
Understanding Lists is the foundation for DataFrames because a DataFrame is essentially a collection of specialized lists (called Series).

A List is a 1D sequence.

A DataFrame is a 2D structure where each column behaves like a list, sharing a common index. If you can manipulate a list, you can understand how to manipulate a column.
