```python
# Task 1/6

start = date(2025, 1, 10)
print(f"Start: {start}")
after_10_days = start + timedelta(days = 10)
print(f"After 10 days: {after_10_days}")

before_14_days = start - timedelta(days = 14)
print(f"Before 14 days: {before_14_days}")

after_3_weeks = start + timedelta(weeks = 3)
print(f"After 3 weeks: {after_3_weeks}")

my_datetime = datetime(2025, 1, 10, 0, 0, 0)

after_5_hours = my_datetime + timedelta(hours = 5)
print(f"The datetime after 5 hours: {after_5_hours}")

after_20_minutes = after_5_hours + timedelta(minutes = 20)
print(f"The datetime after 20 minutes: {after_20_minutes}")

after_45_seconds = after_20_minutes + timedelta(seconds = 45)
print(f"The datetime after 45 seconds: {after_45_seconds}")
```

```python
# Task 2/1

my_df = pd.DataFrame(
    {
        "customer_id" : [1, 2, 3, 4, 5],
        "signup_date" : ["2024-01-10", "2024-02-05", "2024-03-12", "2024-04-01", "2024-05-20"],
        "last_purchase" : ["2024-02-15", "2024-03-01", "2024-03-28", "2024-05-10", "2024-06-25"]
    }
)

print(my_df)
```

```python
my_df["signup_date"] = pd.to_datetime(my_df["signup_date"])
my_df["last_purchase"] = pd.to_datetime(my_df["last_purchase"])
```

```python
my_df.dtypes
```

```python
# Task 2/7

my_df["signup_year"] = my_df["signup_date"].dt.year
my_df["signup_month"] = my_df["signup_date"].dt.month
my_df["signup_day"] = my_df["signup_date"].dt.day
my_df["signup_quarter"] = my_df["signup_date"].dt.quarter
my_df["signup_month_name"] = my_df["signup_date"].dt.month_name()
my_df["signup_day_name"] = my_df["signup_date"].dt.day_name()
print(my_df.head())
print(my_df.dtypes)
```

```python
# Task 2/8

my_df["signup_dow"] = my_df["signup_date"].dt.day_of_week
my_df["is_weekend"] = my_df["signup_date"].dt.day_of_week.isin([5, 6]) # becasue in pandas 5 - saturday, 6 - sunday
print(my_df.head())
```

```python
# Task 2/9

my_df["days_between_signup_and_purchase"] = (my_df["last_purchase"] - my_df["signup_date"]).dt.days
print(my_df.head())
print(my_df.dtypes)
```

```python
# Task 2/15


some_date = pd.Timestamp("2024-08-18")
print(f"Original date: {some_date}")

month_period = some_date.to_period("M")

beginning = month_period.start_time
end = month_period.end_time
print(f"Beginning of month: {beginning}")
print(f"End of month: {end}")

beginning_as_date = beginning.date()
end_as_date = end.date()

print(f"Beginning of month as date only: {beginning_as_date}")
print(f"End of month as date only: {end_as_date}")
```
